```python
import os
from pathlib import Path

DEPLOY_REPO_URL = "https://github.com/RichardPovinelli/SGA_workshop_deploy_test.git"
DEPLOY_REPO_DIR = "SGA_workshop_deploy_test"

if not Path(DEPLOY_REPO_DIR).exists():
    os.system(f"git clone {DEPLOY_REPO_URL}")
os.chdir(DEPLOY_REPO_DIR)
print(f"Working directory: {Path.cwd()}")
```

```python
import subprocess, sys

subprocess.check_call([sys.executable, "-m", "pip", "install", "-r", "requirements.txt", "-q"])
print("Requirements installed.")
```

```python
import sys
from pathlib import Path

repo_root = Path.cwd()
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

from src.runtime_assets import (
    get_dataset_asset_descriptor,
    install_dataset_from_file,
    load_install_state,
)
from src.config import DATASET_INSTALL_STATE_PATH, DATASET_MANIFEST_FILENAME

descriptor = get_dataset_asset_descriptor()
if load_install_state(DATASET_INSTALL_STATE_PATH) is not None:
    print("Dataset already installed — skipping.")
else:
    install_dataset_from_file(
        Path("data") / descriptor.filename,
        source_manifest_path=Path("data") / DATASET_MANIFEST_FILENAME,
        requested_release_tag=descriptor.release_tag,
        requested_filename=descriptor.filename,
    )
    print("Dataset installed.")
```

```python
from src.runtime_assets import validate_installed_dataset

df = validate_installed_dataset()
print(f"Dataset ready: {len(df):,} rows, zones={sorted(df['zone'].unique().tolist())}")
```
