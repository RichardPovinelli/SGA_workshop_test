# 03 Multi Horizon

Run direct XGBoost across horizons 1 through 7 and inspect degradation.

```python
from pathlib import Path
import sys

repo_root = Path.cwd()
if repo_root.name == "notebooks":
    repo_root = repo_root.parent

if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))
```

```python
import matplotlib

matplotlib.use("Agg")
import matplotlib.pyplot as plt
import pandas as pd

from src.data.load import load_dataset
from src.models.xgb import evaluate_xgb_for_all_zones

df = load_dataset()
all_results = []
for horizon in range(1, 8):
    _, horizon_results = evaluate_xgb_for_all_zones(df, horizon)
    all_results.append(horizon_results)

results = pd.concat(all_results, ignore_index=True)
results
```

```python
summary = results.groupby("horizon", as_index=False)["rmse"].mean()
ax = summary.plot(x="horizon", y="rmse", marker="o", title="Average RMSE by Horizon")
ax.set_ylabel("RMSE")
plt.tight_layout()
```
